```python
dtfull = dt.fillna(0)
```

```python
dtfull.isna().sum()
```

```python
dtfullclean = dtfull.drop(['Identyfikator_PS', 'Vidomchyj_kod_PS', 'Rajon_richkovoho_basejnu', 'Date', 'Nazva_laboratoriyi', 'Data_sposterezhen'], axis = 1)
```

```python
from sklearn import preprocessing
```

```python
collist = list(dtfullclean)
collist = collist[3:]
collist
```

```python
#Normalize all columns
dtfullclean[collist] = dtfullclean[collist].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
dtg = dtfullclean.groupby("Nazva_PS", as_index=False).mean()
```

```python
dtg_clusterpoints = dtg[["Nazva_PS", "Shyrota", "Dovhota"]]
dtg_plain = dtg.drop(["Nazva_PS", "Shyrota", "Dovhota"], axis = 1)
dtg_plain.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
```

```python
dtg["Cluster"] = model.fit_predict(dtg_plain)
```

```python
dtg.groupby("Cluster").count()
```

```python
sns.clustermap(dtg_plain)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(dtg_plain, 'ward')
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

c, coph_dists = cophenet(Z, pdist(dtg_plain))
c
```

```python
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z, 6, criterion='maxclust')
len(clusters)
```

```python
dtg_clusterpoints_kmeans = dtg_clusterpoints
dtg_clusterpoints_kmeans["Cluster"] = model.fit_predict(dtg_plain)
dtg_clusterpoints_kmeans.to_csv("KMeans_clusterization4.csv")
```

```python
dtg_clusterpoints_hierarchical = dtg_clusterpoints
dtg_clusterpoints_hierarchical["Cluster"] = clusters
dtg_clusterpoints_hierarchical.head()
dtg_clusterpoints_hierarchical.to_csv("Hclust_clusterization.csv")
```
